```python
import pandas as pd
from google.colab import files

# Upload the CSV file
uploaded = files.upload()

# Get the file name
file_name = list(uploaded.keys())[0]

# Read the CSV file into a DataFrame
df = pd.read_csv(file_name)
```

```python
# Display the DataFrame
df.head()
```

```python
df.dtypes
```

```python
df['lpep_pickup_datetime'] = pd.to_datetime(df['lpep_pickup_datetime'])
df['lpep_dropoff_datetime'] = pd.to_datetime(df['lpep_dropoff_datetime'])
```

```python
#DIM_date table

DIM_date = df[['lpep_pickup_datetime', 'lpep_dropoff_datetime']]
DIM_date = DIM_date.drop_duplicates()

#defining pickup datetime details
DIM_date['pick_day'] = df['lpep_pickup_datetime'].dt.day
DIM_date['pick_month'] = df['lpep_pickup_datetime'].dt.month
DIM_date['pick_hour'] = df['lpep_pickup_datetime'].dt.hour
DIM_date['pick_weekday'] = df['lpep_pickup_datetime'].dt.weekday
DIM_date['pick_year'] = df['lpep_pickup_datetime'].dt.year

#defining dropoff datetime details
DIM_date['drop_day'] = df['lpep_dropoff_datetime'].dt.day
DIM_date['drop_month'] = df['lpep_dropoff_datetime'].dt.month
DIM_date['drop_hour'] = df['lpep_dropoff_datetime'].dt.hour
DIM_date['drop_weekday'] = df['lpep_dropoff_datetime'].dt.weekday
DIM_date['drop_year'] = df['lpep_dropoff_datetime'].dt.year

#to create the primary key column
DIM_date['date_id'] = DIM_date.index

DIM_date = DIM_date[['date_id','lpep_pickup_datetime','lpep_dropoff_datetime','pick_day','pick_hour','pick_weekday','pick_month','pick_year','drop_day','drop_hour','drop_weekday','drop_month','drop_year']]
DIM_date.head()
```

```python
#DIM_passenger table

DIM_passenger = df[['passenger_count']].drop_duplicates().reset_index(drop=True)

#to create the primary key column
DIM_passenger['passenger_id'] = DIM_passenger.index
#defining columns
DIM_passenger['passenger_count'] = df['passenger_count']

#re-arranging the columns
DIM_passenger = DIM_passenger[['passenger_id','passenger_count']]
DIM_passenger.head()
```

```python
#DIM_trip_distance table

DIM_trip_distance = df[['trip_distance']].drop_duplicates().reset_index(drop=True)

#to create the primary key column
DIM_trip_distance['trip_distance_id'] = DIM_trip_distance.index
#defining columns
DIM_passenger['trip_distance'] = df['trip_distance']

#re-arranging the columns
DIM_trip_distance = DIM_trip_distance[['trip_distance_id','trip_distance']]
DIM_trip_distance.head()
```

```python
#DIM_payment_type table
payment_type_name = {
1 : "Credit card",
2 :"Cash",
3 :"'No charge",
4 :"Dispute",
5 :"Unknown",
6 :"Voided trip"
}

DIM_payment_type = df[['payment_type']].drop_duplicates().reset_index(drop=True)

#to create the primary key column
DIM_payment_type['payment_type_id'] = DIM_payment_type.index
#defining columns
DIM_payment_type['payment_type'] = df['payment_type']
DIM_payment_type['payment_type_name'] = DIM_payment_type['payment_type'].map(payment_type_name)
#re-arranging the columns
DIM_payment_type = DIM_payment_type[['payment_type_id','payment_type','payment_type_name']]
DIM_payment_type.head()
```

```python
#DIM_trip_type table

DIM_trip_type = df[['trip_type']].drop_duplicates().reset_index(drop=True)

#to create the primary key column
DIM_trip_type['trip_type_id'] = DIM_trip_type.index
#defining columns
DIM_trip_type['trip_type'] = df['trip_type']

#re-arranging the columns
DIM_trip_type = DIM_trip_type[['trip_type_id','trip_type']]
DIM_trip_type.head()
```

```python
#DIM_rate_code table
rate_code_type = {
1 : "Standard rate",
2 :"JFK",
3 :"Newark",
4 :"Nassau or Westchester",
5 :"Negotiated fare",
6 :"Group ride"
}


DIM_rate_code = df[['RatecodeID']].drop_duplicates().reset_index(drop=True)

#to create the primary key column
DIM_rate_code['rate_code_id'] = DIM_rate_code.index
#defining columns
#DIM_rate_code['rate_code'] = df['RatecodeID']
DIM_rate_code['rate_code_name'] = DIM_rate_code['RatecodeID'].map(rate_code_type)
#re-arranging the columns
DIM_rate_code = DIM_rate_code[['rate_code_id','RatecodeID','rate_code_name']]
DIM_rate_code.head()


```

```python
#creating the fct table
fact_yellow_taxi = df.merge(DIM_trip_distance, on='trip_distance') \
    .merge(DIM_date, on=['lpep_pickup_datetime', 'lpep_dropoff_datetime']) \
    .merge(DIM_passenger, on='passenger_count') \
    .merge(DIM_payment_type, on='payment_type') \
    .merge(DIM_trip_type, on='trip_type') \
    .merge(DIM_rate_code, on='RatecodeID') \
    [['VendorID', 'date_id', 'trip_distance_id', 'passenger_id', 'payment_type_id', 'trip_type_id', 'rate_code_id',
      'store_and_fwd_flag', 'PULocationID', 'DOLocationID', 'fare_amount', 'extra', 'mta_tax', 'tip_amount',
      'tolls_amount', 'ehail_fee', 'improvement_surcharge', 'total_amount', 'congestion_surcharge']]

fact_yellow_taxi.head()
```
